```python
class Phone:
    def __init__(self, brand, model, color, price):
        self.brand = brand
        self.model = model
        self.color = color
        self.price = price

    def display_info(self):
        print(f'{self.brand} {self.model}, {self.color} color: ${self.price}')
```

```python
iphone1 = Phone('Apple', 'iPhone 14 Pro', 'Silver', 999)
iphone1.display_info()

iphone2 = Phone('Apple', 'iPhone 14', 'Black', 799)
iphone2.display_info()

xiaomi1 = Phone('Xiaomi', 'Redmi Note 11', 'Blue', 599)
xiaomi1.display_info()
```

```python
class Store:
    def __init__(self, name):
        self.name = name
        self.fruits = ['Apple', 'Banana', 'Orange', 'Grapes']

    def display_fruits(self):
        print(f'Fruits available in {self.name}:')
        for fruit in self.fruits:
            print(f'- {fruit}')

    def add_fruit(self, fruit_name):
        self.fruits.append(fruit_name)
        print(f'{fruit_name} has been added to the store.')

    def remove_fruit(self, fruit_name):
        # validation
        if fruit_name not in self.fruits:
            print(f'{fruit_name} is not available in the store.')
            return
        
        # main logic
        self.fruits.remove(fruit_name)
        print(f'{fruit_name} has been removed from the store.')
```

```python
win_mart = Store('Win Mart')
win_mart.display_fruits()
```

```python
win_mart.add_fruit('Mango')
win_mart.add_fruit('Pineapple')
win_mart.display_fruits()
```

```python
win_mart.remove_fruit('Banana')
win_mart.display_fruits()
```
