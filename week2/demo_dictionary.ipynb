```python
# declare dictionary {}
en_vn = {
    "hello": "xin chào",    # key: value where key is unique, value can be duplicate
    "goodbye": "tạm biệt",
    "thank you": "cảm ơn"
    }
```

```python
print(en_vn)
```

```python
print(en_vn['hello'])  # access value by key
```

```python
print(en_vn['hi'])  # access value by invalid key will raise KeyError
```

```python
# add new key-value pair
en_vn['banana'] = 'quả chuối'
print(en_vn)
```

```python
# edit a current key-value pair
en_vn['banana'] = 'chuối'
print(en_vn)
```

```python
# delete a key-value pair
del en_vn['banana']
print(en_vn)
```

```python
# another way to delete dictionary
en_vn.pop('goodbye')  # remove key-value pair and return value
print(en_vn)
```

```python
keys = list(en_vn.keys())  # get all keys
print(keys)
```

```python
values = list(en_vn.values())  # get all values
print(values)
```

```python
items = list(en_vn.items())  # get all key-value pairs
print(items)
```

```python
for en, vn in en_vn.items():
    print(f'{en} -> {vn}')
```
